```python
# from the above statistic it can be concluded that the quantity column contains negative values 
# lets replace the negative values with absolute values 

negative_values = df[df['Quantity'] < 0] # filter rows with negative values 

# Replace negative values with their positive counterparts
negative_values['Quantity'] = negative_values['Quantity'].abs()

# Add the converted positive values back to the original data
df.loc[negative_values.index, 'Quantity'] = negative_values['Quantity']
```

```python
# check to confirm if the data contains negative values 
df.describe()
```

```python
# Calculate the IQR (Interquartile Range) for the 'Quantity' column
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame without outliers
df_no_outliers = df[(df['Quantity'] >= lower_bound) & (df['Quantity'] <= upper_bound)]
```

```python
# check the basic statistic after the removal of outliers 
df_no_outliers.describe()
```

```python
# display the boxplot after the removal of outliers 
import matplotlib.pyplot as plt 
plt.boxplot(df_no_outliers['Quantity'])
plt.show()
```

```python
# assign df_no_outliers to df_1
df_1 = df_no_outliers
```

```python
# Display unique values in a categorical column
df_1['Country'].unique()
```

```python
# convert InvoiceDate to Datetime 
df_1['InvoiceDate'] = pd.to_datetime(df_1['InvoiceDate'])
```

```python
# Create a 'Month' column based on the 'InvoiceDate' column
df_1['Month'] = df_1['InvoiceDate'].dt.to_period('M')
```

```python
df_1
```

```python
df_1.reset_index(drop=True)
```

```python
# Aggregate the data by month and calculate the sum of 'Quantity' for each month
monthly_sales = df_1.groupby('Month')['Quantity'].sum()
monthly_sales
```

```python
# Convert the 'monthly_sales' Series to a DataFrame
monthly_sales_df = monthly_sales.reset_index()
monthly_sales_df
```

```python
# Rename the columns
monthly_sales_df.columns = ['Month', 'Total Quantity']
monthly_sales_df
```

```python
# plot the monthly sales data 
# Convert the 'Month' values to strings
monthly_sales_df['Month'] = monthly_sales_df['Month'].dt.strftime('%Y-%m')

# Plot the monthly sales data
plt.figure(figsize=(12, 6))  # Set the figure size

plt.plot(monthly_sales_df['Month'], monthly_sales_df['Total Quantity'], marker='o', linestyle='-', color='b')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Calculate 'TotalSpent' as Quantity * UnitPrice
df_1['TotalSpent'] = df_1['Quantity'] * df_1['UnitPrice']
```

```python
# display the data
df_1
```

```python
# Count of unique values
df_1['Country'].value_counts()
```

```python
# Group the data by 'Country' and calculate the sum of 'TotalSpent' for each country
country_totals = df_1.groupby('Country')['TotalSpent'].sum()
country_totals
```

```python
country_totals.reset_index()
```

```python
# import funcformatter 
from matplotlib.ticker import FuncFormatter

# Create a bar plot
plt.figure(figsize=(12, 6))  # Set the figure size

# Function to format y-axis labels with currency symbol and comma separators
def currency_formatter(x, pos):
    return f"${x:,.0f}"

# Use the formatter for the y-axis
formatter = FuncFormatter(currency_formatter)
plt.gca().yaxis.set_major_formatter(formatter)

country_totals.plot(kind='bar')
plt.title('Total Amount Spent by Country')
plt.xlabel('Country')
plt.ylabel('Total Amount Spent')

# Rotate x-axis labels for better readability (adjust as needed)
plt.xticks(rotation=90)


plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Group the data by 'StockCode' and calculate the sum of 'Quantity' for each product
product_sales = df_1.groupby('StockCode')['Quantity'].sum()

# Find the product with the highest total quantity sold
best_selling_product = product_sales.idxmax()  # Get the index (StockCode) of the product

# Get the actual product name or description
best_selling_product_name = df_1[df_1['StockCode'] == best_selling_product]['Description'].iloc[0]

print(f"The best-selling product is {best_selling_product_name} (StockCode: {best_selling_product})")
```

```python
# display the top selling products in descending order 
product_sales = df_1.groupby('StockCode')['Quantity'].sum()

# Sort the products by total quantity sold in descending order
top_selling_products = product_sales.sort_values(ascending=False)
top_selling_products
```

```python
# Display the top N selling products (e.g., top 10)
top_n = 10  # Change this to the desired number of top-selling products

print(f"Top {top_n} Selling Products:")
for i, (stock_code, total_quantity) in enumerate(top_selling_products.head(top_n).items()):
    product_name = df_1[df_1['StockCode'] == stock_code]['Description'].iloc[0]
    print(f"{i + 1}. {product_name}, (StockCode: {stock_code}) - Total Quantity Sold: {total_quantity}")
```

```python
# Group the data by 'CustomerID' and calculate the sum of 'TotalSpent' for each customer
customer_spending = df_1.groupby('CustomerID')['TotalSpent'].sum()

# Find the customer with the highest total spending
most_valuable_customer = customer_spending.idxmax()  # Get the index (CustomerID) of the customer

print(f"The most valuable customer is  (CustomerID: {most_valuable_customer})")
```

```python
# Group the data by 'CustomerID' and calculate the sum of 'TotalSpent' for each customer
customer_spending = df_1.groupby('CustomerID')['TotalSpent'].sum()

# Sort the customers by total spending in descending order
top_customers = customer_spending.sort_values(ascending=False)

# Display the top N customers 
top_n = 10  # Change this to the desired number of top customers

print(f"Top {top_n} Customers:")
for i, (customer_id, total_spending) in enumerate(top_customers.head(top_n).items()):
    print(f"{i + 1}. Customer ID: {customer_id} - Total Spending: ${total_spending:,.2f}")
```
